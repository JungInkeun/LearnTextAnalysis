```python
print("지금 치킨을 먹고 있습니다.")
term = "목이 마르다"

if term == '목이 마르다':
    print("조건 만족")
    print("치킨무를 먹는다.")

else:
    print("조건 불만족")
    print("계속 치킨을 먹는다.")
```

```python
print("상황 : 지금 야식을 먹고 싶다...")

term1 = '치킨'
term2 = '튀김'
term3 = '라면'
term4 = '결정못함'

choice = term1

if choice == term1:
    print("치킨을 선택했습니다.")
    print("A치킨집에 전화합니다.")

elif choice == term2:
    print("튀김을 선택했습니다.")
    print("B분식집에 전화합니다.")

elif choice == term3:
    print("라면을 선택했습니다.")
    print("편의점에 갑니다.")

else:
    print("선택하지 않았습니다.\n그냥 잡니다.")
```
